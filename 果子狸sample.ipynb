<h2> 1) Timeline Overview -- Daily </h2>

```python
date_list1 = []

for each in csv1['发布时间']:
    date_list1.append(each)
    
date_list1
```

```python
# 去掉2020，以确保后面的visualization的简洁

date_list1_clean = []

for each in date_list1:
    cancel = each.replace('2020/','')
    date_list1_clean.append(cancel)
    
date_list1_clean
```

```python
from datetime import datetime

# 抽出日期进行转化

dates = []

for day in date_list1_clean:
    current_date = datetime.strptime(day, '%m/%d %H:%M')
    dates.append(current_date)

dates
```

```python
dates1 = []
    
for each in dates:
    dates1.append(each.strftime('%d/%m'))
    
dates1
```

<h2> 2) Hashtag </h2>
<h3> A. Hashtag Frequency </h3>

```python
import re

ht1_list = []

for line in csv1['微博正文']:
    line_s = line.split('\n')
    
    for i in range(0, len(line_s)):
        if i == 0:
            lst = re.findall(r'(#.+?#)', line_s[i])
    
            if lst is not None:
                for each in lst:
                    ht1_list.append(each)
        
        if i == 2:
            lst = re.findall(r'(#.+?#)', line_s[i])
            
            if lst is not None:
                for each in lst:
                    ht1_list.append(each)
                    
ht1_list
#成功
```

<h3> B. User-Hashtag Network </h3>

```python
import re

ht1_newlist = []

for line in csv1['微博正文']:
    line_s = line.split('\n')
    
    q = []
    p = ''
    
    for i in range(0, len(line_s)):
        if i == 0:
            lst = re.findall(r'(#.+?#)', line_s[i])
            
            if lst is not None:
                for each in lst:
                    q.append(each)
                    
        if i == 2:
            lst = re.findall(r'(#.+?#)', line_s[i])
            
            if lst is not None:
                for each in lst:
                    q.append(each)
                
    for i in range(len(q)):
        if i == (len(q)-1):
            p += q[i]
        else:
            p += q[i] + ', '
#     print(p)
    
    if p != '':
        ht1_newlist.append(p)
        
ht1_newlist
```

```python
lst = []
for i in range(len(ht1_newlist)):
    lst.append('User 1')
    
user_hashtag_net1 = pd.DataFrame({'Source': lst,
                                 'Target': ht1_newlist})
user_hashtag_net1
```

<h2> 2) @Mention </h2>
<h3> A. Mention Frequency </h3>

```python
# find out the users/friends online @果子狸7777 always repost from
# 找出所有mentions

import re

mt1_list = []

for line in csv1['微博正文']:
    line_s = line.split('\n')
    
    for i in range(0, len(line_s)):
        if i == 0:
            lst = re.findall(r'(@.+?):', line_s[i])
#             print(lst)

            if lst is not None:
                for each in lst:
                    mt1_list.append(each)

        if i == 1:
            lst = re.findall(r'原始用户: (.+)',line_s[i])
#             print(lst)

            if lst is not None:
                for each in lst:
                    mt1_list.append('@' + each)

        if i == 2:
            lst = re.findall(r'(@.+?) ', line_s[i])
#             print(lst)

            if lst is not None:
                for each in lst:
                    mt1_list.append(each)
                    
mt1_list
```

```python
average = len(mt1_list) / len(csv1)
print('User 1 mentions ' + str(average) + ' users per post in average.')
```

```python
mt1_fre = {}

for user in mt1_list:
    mt1_fre[user] = mt1_fre.get(user, 0) +1
    
mt1_fre
```

<h3> B. User-User Newtwork </h3>

```python
import re

mt1_newlist = []

for line in csv1['微博正文']:
    line_s = line.split('\n')
    
    q = [] # 针对每一行给出一个list装载该行所有@mention，且每一次循环q都会清洗一次
    p = ''
    
    for i in range(0, len(line_s)):  
        if i == 0:
            lst = re.findall(r'(@.+?):', line_s[i])
#             print(lst)

            if lst is not None:
                for each in lst:
                    q.append(each)

        if i == 1:
            lst = re.findall(r'原始用户: (.+)',line_s[i])
#             print(lst)

            if lst is not None:
                for each in lst:
                   q.append('@' + each)

        if i == 2:
            lst = re.findall(r'(@.+?) ', line_s[i])
#             print(lst)

            if lst is not None:
                for each in lst:
                   q.append(each)
                    
#     print(q) -- 包括没有@mention的行
        
    # 对于
    for i in range(len(q)):
        if i == (len(q)-1):
            p += q[i]
        else:
            p += q[i] + ', '
#     print(p)
    
    # 除去没有@mention的行
    if p != '':
        mt1_newlist.append(p)
        
mt1_newlist
```

```python
file_lst = ['11_20_May.csv', '21_31_May.csv', '1_10_June.csv', '11_20_June.csv', '21_30_June.csv']
period = ['11-20 May', '21-31 May', '1-10 June', '11-20 June', '21-30 June']
lst = []
df = pd.concat([ht_df], axis=1)

for t in range(5):  
    open = pd.read_csv(file_lst[t])
    
    # 取出所有hashtag
    import re

    ht_list = []

    for line in open['微博正文']: #需要修改
        line_s = line.split('\n')

        for i in range(0, len(line_s)):
            if i == 0:
                lst = re.findall(r'(#.+?#)', line_s[i])

                if lst is not None:
                    for each in lst:
                        ht_list.append(each)

            if i == 2:
                lst = re.findall(r'(#.+?#)', line_s[i])

                if lst is not None:
                    for each in lst:
                        ht_list.append(each)
#     ht_list

    # 做成dictionary
    ht_fre = {}

    for tag in ht_list:
        ht_fre[tag] = ht_fre.get(tag, 0) +1
    print(len(ht_fre))
#     ht_fre    
    
    # 转化成df
    ht_df_ = pd.DataFrame({period[t]: [user for user in ht_fre.keys()],
                      'Count': [count for count in ht_fre.values()]})
    ht_df_ = ht_df_.sort_values('Count', ascending=False).reset_index(drop=True)
#     print(len(ht_df_))

    df = pd.concat([df,ht_df_], axis=1)
```
